## 1. Set the environment variables

```python
import os

# Azure Content Understanding env variables
os.environ["AZURE_CONTENT_UNDERSTANDING_ENDPOINT"] = "YOUR_AZURE_CONTENT_UNDERSTANDING_ENDPOINT_HERE"
os.environ["AZURE_CONTENT_UNDERSTANDING_KEY"] = "YOUR_AZURE_CONTENT_UNDERSTANDING_KEY_HERE"

# Openlayer env variables
os.environ["OPENLAYER_API_KEY"] = "YOUR_OPENLAYER_API_KEY_HERE"
os.environ["OPENLAYER_INFERENCE_PIPELINE_ID"] = "YOUR_OPENLAYER_INFERENCE_PIPELINE_ID_HERE"
```

## 2. Import the `trace_azure_content_understanding` function and create the client

```python
from azure.core.credentials import AzureKeyCredential
from azure.ai.contentunderstanding import ContentUnderstandingClient
from azure.ai.contentunderstanding.models import AnalysisInput

from openlayer.lib import configure, trace_azure_content_understanding

# Configure if you want to upload documents to Openlayer storage
configure(
    attachment_upload_enabled=True,   # upload binary/file attachments
    url_upload_enabled=True,          # also download & re-upload external URLs
)

client = trace_azure_content_understanding(
    ContentUnderstandingClient(
        endpoint=os.environ.get("AZURE_CONTENT_UNDERSTANDING_ENDPOINT"),
        credential=AzureKeyCredential(os.environ.get("AZURE_CONTENT_UNDERSTANDING_KEY")),
        api_version="2025-11-01",
    )
)
```

## 3. Use your traced client normally

That's it! Now you can continue using your Azure Content Understanding client normally. The data is automatically published to Openlayer and you can start creating tests around it!

```python
analyzer_id = "prebuilt-read"
url = "https://contentunderstanding.ai.azure.com/assets/prebuilt/read_healthcare.png"

poller = client.begin_analyze(
    analyzer_id=analyzer_id,
    inputs=[AnalysisInput(url=url)],
)
result = poller.result()
```
